```python
# Instalacion de dependencias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# === Carga de datos ===
df = pd.read_csv("evaluation_results.csv")
```

```python
# === Agrupar métricas por proveedor ===
metrics = df.groupby("provider").agg({
    "precision_at_k": "mean",
    "exact_match": "mean",
    "cosine_similarity": "mean",
    "citation_presence": "mean"
}).round(4)

print("=== Resumen de la evaluación ===")
display(metrics)
```

# Comparativa de Exact Match y Similitud Coseno

Esta sección del texto se enfoca en la evaluación de las respuestas generadas por los Modelos de Lenguaje Grande (LLM), comparando dos métricas: la Coincidencia Exacta (Exact Match o EM) y la Similitud del Coseno (Cosine Similarity).

Problema de la Coincidencia Exacta (EM): Se señala que las respuestas de los LLM no coinciden exactamente con la respuesta "correcta" predefinida en el gold_set (conjunto de referencia o verdad fundamental). Como resultado, la puntuación de  para todas las respuestas es 0.

```python


fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Exact Match promedio por provider
sns.barplot(data=df, x="provider", y="exact_match", ax=axes[0],errorbar=None)
axes[0].set_title("Exact Match promedio")
axes[0].set_ylabel("Exact Match")

# Cosine Similarity promedio por provider
sns.barplot(data=df, x="provider", y="cosine_similarity", ax=axes[1], errorbar=None)
axes[1].set_title("Similitud coseno promedio")
axes[1].set_ylabel("Cosine Similarity")

plt.show()
```
